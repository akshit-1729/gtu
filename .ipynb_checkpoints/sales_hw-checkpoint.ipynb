```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


sales_data = np.loadtxt("Sales.csv", delimiter = ',', skiprows = 1, usecols = range(1,8))
months = np.loadtxt("Sales.csv", delimiter = ',', skiprows = 1, usecols = 0, dtype=str)
sales_data
```

```python
sales_data.shape
total_profit = np.arange(12)
for i in range(12):
    total_profit[i] = sales_data[i][6]
total_profit

yval = np.arange(0,340000,30000)

plt.rcParams["figure.figsize"] = [12,5]
xval = np.arange(1,13)
plt.plot(xval,total_profit)
plt.xticks(xval,months)
plt.yticks(yval)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [10,5]
xval = np.arange(1,13)

plt.plot(xval, total_profit, marker = 'o', mec = 'y', mfc = 'y', linestyle = 'dotted', lw = '5', color = 'blue')
plt.xticks(xval)
plt.xlabel('Month Number')
plt.yticks(yval)
plt.ylabel('Item Sold Units Number')
plt.legend(['total units sold over months'], loc = "upper right")
plt.show()
```

```python
A = sales_data[0].sum()
B = sales_data[1].sum()
C = sales_data[2].sum()
D = sales_data[3].sum()
E = sales_data[4].sum()

y = np.array([A,B,C,D,E])
l = ["A", "B", "C", "D", "E"]
plt.pie(y, labels = l)
plt.show()
```
