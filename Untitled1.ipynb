```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pylab as pltt


```

```python
train = pd.read_csv('train.csv')
train
testt = pd.read_csv('test.csv')
```

```python
test_id = train['id']
x = train.drop(columns=['id', 'clarity', 'depth'])
x = pd.get_dummies(x)
y = x['price']
x = x.drop(columns= 'price')
x.info()
```

```python
test = pd.read_csv('test.csv')
test_id = test['id']

x1 = test.drop(columns=['id','clarity','depth'])
x1 = pd.get_dummies(x1)

x1
```

```python
fig, ax = plt.subplots(3, 1, figsize=(25, 25), dpi=100)
ax = ax.flatten()
for i, ft in enumerate(['cut', 'color', 'clarity']):
    sns.histplot(
        data=train,
        x="price", hue=ft,
        multiple="stack",
        palette="dark:yellow",
        edgecolor=".3",
        linewidth=.5,
        log_scale=True,
        ax=ax[i],
        
    )
fig.suptitle(f'Categorical Features vs Price\n\n\n', ha='center',  fontweight='bold', fontsize=25)
plt.tight_layout()
plt.show()
```

```python
cuts = ['Fair','Good','Very Good','Premium','Ideal']
colors = ['J','I','H','G','F','E','D']
clars = ['I3','I2','I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF','FL']

cut_dict = {cuts[i]:i for i in range(len(cuts))}
color_dict = {colors[i]:i for i in range(len(colors))}
clar_dict = {clars[i]:i for i in range(len(clars))}

train.cut = train.cut.map(cut_dict)
train.color = train.color.map(color_dict)
train.clarity = train.clarity.map(clar_dict)


testt.cut = testt.cut.map(cut_dict)
testt.color = testt.color.map(color_dict)
testt.clarity = testt.clarity.map(clar_dict)

testt
```

```python
import statsmodels.api as sm

train['vol'] = train['x']*train['y']*train['z']
train['dpthpervol'] = train['vol']/train['depth']
train['depthpertable'] = train['depth']/train['table']
x = train[['price', 'carat', 'depth', 'table', 'x', 'y' , 'z',  'vol', 'dpthpervol', 'depthpertable', 'color', 'cut', 'clarity']]
y = train['price']
x = x.drop(columns=['price','table','depthpertable','x','y','z','depth'])


#
testt['vol'] = testt['x']*testt['y']*testt['z']
testt['dpthpervol'] = testt['vol']/testt['depth']
testt['depthpertable'] = testt['depth']/testt['table']
#
data_train, data_test, y_train, y_test = tts(x,y,train_size = 0.9, test_size = 0.1, random_state = 100)

data_train_sm = sm.add_constant(data_train)
data_train_sm.corr()

```

```python
lr = sm.OLS(y_train, data_train_sm).fit()
lr.params

```

```python
y_train_pred = lr.predict(data_train_sm)
res = (y_train - y_train_pred)
res
```

```python
data_test_sm = sm.add_constant(data_test)

y_test_pred = lr.predict(data_test_sm)
y_test_pred
```

```python
from sklearn.metrics import r2_score

# Checking the R-squared value
r_squared = r2_score(y_test, y_test_pred)
r_squared
```
